we are going to analyze the cryptocurrency data from Kaggle. The data is available publicly here https://www.kaggle.com/jackml/cryptocurrency-historical-data/data . In the entire dataset there are top 50 cryptocurrency's data is available. 

Out of all data we are going to analyze the data of Bitcoin, which is the top number cryptocurrency right now. 

We will plot diffferent values of data and will see bitcoin growth over the time. The data is from 2017 to 2013. The dataset includes High,Low,Open,Close,Volume,MarketCap over these years. 

Now let's dig deep into data. We will start by importing libraries in python.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

Now we will read bitcoin.csv file using pandas library. Here the dates are not takenas timestamp,so we will convert them into timestamp using parse function as below.

You can see first five enteries of dataset.

```python
def parse(x):
	return datetime.strptime(x, '%d/%m/%Y')
df=pd.read_csv('bitcoin.csv',delimiter=';',parse_dates=['Date'],date_parser=parse)
df.head()
```

```python
df.set_index(['Date'],inplace=True)
df.head()
```

Now let's plot open and Close prices of the day with respected date and time. 

Here we can see that in 2014 the price went to up bu then in 2015 and 2016 it setteled to low. But because of recent activities in market, the prices are up in 2017 with significant increase. The increment is about 4 times the previous years.

```python
df.plot(figsize=(15,15),x=df.index,y=['Open','Close'],title='Plot of for Open and Close price for Bitcoin')
```

Now let's see the variation in High  and low over the years. The variation is quite same as the open and Close. in 2017 High and low also raises up tp 4 times the previous years.

```python
df.plot(figsize=(15,15),x=df.index,y=['High','Low'],title='Plot of High and Low for Bitcoin')
```

```python
df.plot(figsize=(15,15),x=df.index,y=['Volume','MarketCap'],title='Plot of Volume and MarketCap of Bitcoin')
```

```python
df.plot(figsize=(15,15),title='Plot of Bitcoin',subplots=True)
```
