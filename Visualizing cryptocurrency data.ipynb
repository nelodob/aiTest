In this simple Ipython notebook we will visualize the data of different cryptocurrencies. Today, the market value of each crypto currency is growing rapidly. We will see from which time duration this rise become more dominant. 

First we will import some useful packeges to visualize the data for top five currencies. The name of currencies are as follows:
1. Bitcoin
2. Ethereum
3. Litecoin
4. Bitcoin Cach
5. Ripple

The data is collected from https://coinmarketcap.com/ . The data is publicly available and we have just gethered it for only five currencies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#loading the dataset
df=pd.read_csv('CryptoCurrencyData1.csv')
df.head()
```

As you can see the first five rows of the data. The dataset contain name of currency, date, open price,high price,low price,close price, volume and marketcap value of that currency on that particular day. 

The data for each currency starts at different date as some currency has been launched few years back. The oldest is Bitcoin and you can see the data up to 17 January 2018 for each of the currency.

```python
df['Date']=pd.to_datetime(df['Date'],format='%m/%d/%Y')
```

```python
df.set_index(['Date'],inplace=True)
df.head()
```

```python
df['MovingAvg']=pd.rolling_mean(df['Volume'],100)
df['MovingAvg'].dropna(inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for group,frame in df.groupby(['CryptoCurrency']):
    frame['Close'].plot(figsize=(15,15),title='Close for different cryptocurrency',ax=ax,label=group)
plt.legend()
```

The above plot is of Close price for each day ion year for each currency. Some of the currency has been lauched after others,so there plot starts at some point later. 

For Bitcoin we can say that the daily close price has raised a lot in recent years. AT the begining no one was really interested in cryptocurrency. But now because of it's decentralized application people are getting atracted towards it. 

The price of Bitcoinn hiked from mid 2017 to till date. At some of the point it was around 19,000 USD.

In the recent days prices of most of cryptocurrency are getting down. Let's see what will happen in near future.

After Bitcoin,Bitcoin cash is at second postion in terms of highest close prices. It is also getting better day by day but it is way behind the Bitcoin.

Ethereum is on the third position in terms of close price. But, the recent growth shows that it can surpass the Bitcoin in near future.

Litecoin is on fourth and the Ripple is at last in terms of close price. 

We will see in future which one gets better and which one worse.

```python
fig, ax = plt.subplots(figsize=(8,6))
for group,frame in df.groupby(['CryptoCurrency']):
    frame['Volume'].plot(figsize=(15,15),title='Volume for different cryptocurrency',ax=ax,label=group)
plt.legend()
```

The above plot is of volume of each crypto currency. 

We can see the volume the Bitcoin is highest right now. While the other currencies are comparable to each other.

Now let's compute the moving average for 100 samples in volume and let's see how that plot looks like. Does it give us any good insight or not

```python
fig, ax = plt.subplots(figsize=(8,6))
for group,frame in df.groupby(['CryptoCurrency']):
    frame['MovingAvg'].plot(figsize=(15,15),title='Moving Average of Volume for different cryptocurrency',ax=ax,label=group)
plt.legend()
```

The above plot is of moving average for each currency. 

Here an interesting pattern can be seen. 

The moving average of volume for Bitcoin is getting increased while for other it is decreasing.

Ethereum is on second number after the Bitcoin. While the third is Bitcoin cash, fourth is Litecoin and the fifth is Ripple. 

It is matter of time which one rises to top and which one setteles to botom 

```python
fig, ax = plt.subplots(figsize=(8,6))
for group,frame in df.groupby(['CryptoCurrency']):
    frame['Market Cap'].plot(figsize=(15,15),title='Market Cap Value for different cryptocurrency',ax=ax,label=group)
plt.legend()
```

The above plot is of Marketcap value of each currency.

Clearly Bitcoin is leading the market right now. It has shown significant growth in last year. But this year, its value is collapsing.

Ethereum has also shown a rapid growth in last couple of months. It is on the second after the Bitcoin.

But Ripple is approaching Ethereum very soon. It is on third position right now as we can see in plot. It is matter of time that it may take over Ethereum soon.

The Litecoin and Bitcoin cash are on fifth and fourth spot in this plot.
